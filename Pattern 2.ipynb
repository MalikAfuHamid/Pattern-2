````python
baris = int(input("Silakan Masukkan Jumlah Barisnya: "))

#SEGITIGA 1
i = 1
while i <= baris-1: #while pertama (Outer loop)
    j = 1
    while j <= baris: #while kedua (Inner loop)
        if j < i:
            print(' ', end = '  ')
        else:
            print('*', end = '  ')
        j = j + 1 #ini digunakan ketika inner loop j mencapai jumlah yang tidak sesuai dengan 'while j <= baris' sehingga dia akan keluar dari loop...
                  #...kemudian dia akan mengeksekusi baris dibawahnya, yakni i = i+1
    print()
    i = i + 1 #ketika sudah lebih dari 10, maka proses looping selesai dan keluar dari loop karena ex: 11 <= 10 itu tidak sesuai.
    

#SEGITIGA 2
for i in range(1, baris +1):  
    for j in range(1, baris +1):
        if j <= baris - i:
            print(' ', end = '  ')
        else:
            print('*', end = '  ')
    print()
````
